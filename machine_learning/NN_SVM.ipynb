```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_csv('/content/drive/MyDrive/機器學習/bank.csv', delimiter=';', header='infer')
df.y.replace(('yes','no'),(1,0), inplace=True)
df.loan.replace(('yes','no'),(1,0), inplace=True)

lb = LabelEncoder()
df.iloc[:,1] = lb.fit_transform(df.iloc[:,1])
df.iloc[:,2] = lb.fit_transform(df.iloc[:,2])
df.iloc[:,3] = lb.fit_transform(df.iloc[:,3])
df.iloc[:,4] = lb.fit_transform(df.iloc[:,4])
df.iloc[:,6] = lb.fit_transform(df.iloc[:,6])
df.iloc[:,8] = lb.fit_transform(df.iloc[:,8])
df.iloc[:,10] = lb.fit_transform(df.iloc[:,10])
df.iloc[:,15] = lb.fit_transform(df.iloc[:,15])
df
```

```python
x = np.array(df.iloc[:,0:-1]) #x
y = df.iloc[:,-1] #y
print('x:\n',x)
print('y:\n',y)

ss = StandardScaler()
scaled_data = ss.fit_transform(x.reshape(-1,16))
print(scaled_data, scaled_data.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1000)
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train, y_train)
svc.score(x_test, y_test)
```

```python
print(x_train.shape, y_train.shape)
```

```python
import tensorflow as tf

y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
print(y_train, y_train.shape, y_test, y_test.shape)
```

```python
from skmultilearn.problem_transform import BinaryRelevance, ClassifierChain, LabelPowerset
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

classifier = BinaryRelevance(GaussianNB())
classifier.fit(x_train, y_train)
prediction = classifier.predict(x_test)
accuracy_score(y_test, prediction)
```

```python
CCclassifier = ClassifierChain(GaussianNB())
CCclassifier.fit(x_train, y_train)
CCprediction = CCclassifier.predict(x_test)
accuracy_score(y_test, CCprediction)
```

```python
LPclassifier = LabelPowerset(GaussianNB())
LPclassifier.fit(x_train, y_train)
LPprediction = LPclassifier.predict(x_test)
accuracy_score(y_test, LPprediction)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(20, input_dim=16, kernel_initializer='he_uniform', activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(x_train, y_train, epochs=300)
prediction = model.predict(x_test)
```

```python
prediction = prediction.round()
prediction
```
