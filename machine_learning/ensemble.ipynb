```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = pd.read_csv('/content/drive/MyDrive/機器學習/bank.csv', delimiter = ";", header = 'infer')
data.y.replace(('yes','no'),(1,0), inplace=True)
data.loan.replace(('yes','no'),(1,0), inplace=True)
data
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

lb = LabelEncoder()
data.iloc[:,1] = lb.fit_transform(data.iloc[:,1])
data.iloc[:,2] = lb.fit_transform(data.iloc[:,2])
data.iloc[:,3] = lb.fit_transform(data.iloc[:,3])
data.iloc[:,4] = lb.fit_transform(data.iloc[:,4])
data.iloc[:,6] = lb.fit_transform(data.iloc[:,6])
data.iloc[:,8] = lb.fit_transform(data.iloc[:,8])
data.iloc[:,10] = lb.fit_transform(data.iloc[:,10])
data.iloc[:,15] = lb.fit_transform(data.iloc[:,15])
data
```

```python
x = np.array(data.iloc[:,0:-1]) #x
y = data.iloc[:,-1] #y
print('x:\n',x)
print('y:\n',y)

ss = StandardScaler()
scaled_data = ss.fit_transform(x.reshape(-1,16))
print(scaled_data, scaled_data.shape)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=6)
x_pca = pca.fit_transform(x)
print('pca')
print(x_pca)
print('pca variance')
print(x_pca.var(axis=0))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=1000)

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(x_train, y_train)
gnb.score(x_test, y_test)
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train, y_train)
svc.score(x_test, y_test)
```

```python
from sklearn.ensemble import VotingClassifier, RandomForestClassifier

vote = VotingClassifier(estimators=[('lr',lr),('gnb',gnb),('svm',svc)], voting='hard')
vote.fit(x_train, y_train)
vote.score(x_test, y_test)
```

```python
rfc = RandomForestClassifier(n_estimators=50, random_state=1)
rfc.fit(x_train, y_train)
rfc.score(x_test, y_test)
```

```python
vote1 = VotingClassifier(estimators=[('lr',lr),('gnb',gnb),('rf',rfc)], voting='soft')
vote1.fit(x_train, y_train)
vote1.score(x_test, y_test)
```

```python
vote2 = VotingClassifier(estimators=[('lr',lr),('gnb',gnb),('rf',rfc)], voting='soft', weights=[1,1,3])
vote2.fit(x_train, y_train)
vote2.score(x_test, y_test)
```

```python
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier, StackingClassifier

ebc = BaggingClassifier(base_estimator=RandomForestClassifier(), n_estimators=10, max_features=0.8, bootstrap_features=False, max_samples=0.8)
ebc.fit(x_train, y_train)
ebc.score(x_test, y_test)
```

```python
ada = AdaBoostClassifier(base_estimator=RandomForestClassifier(), n_estimators=100, learning_rate=1.0)
ada.fit(x_train, y_train)
ada.score(x_test, y_test)
```

```python
stacking = StackingClassifier(estimators=[('lr',lr),('gnb',gnb),('rf',rfc),('svm',svc)], final_estimator=SVC(), cv=30)
stacking.fit(x_train, y_train)
stacking.score(x_test, y_test)
```
